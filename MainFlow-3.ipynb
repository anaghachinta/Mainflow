```python
import pandas as pd
import matplotlib.pyplot as plt
ds1 = pd.read_csv('EV_Market.csv')  
plt.figure(figsize=(16, 6))
plt.bar(ds1['Brand'], ds1['AccelSec'], color='violet')
plt.xlabel('Brands', family='serif', fontsize=15, labelpad=15)  #labels
plt.ylabel('Acceleration (Sec)', family='serif', fontsize=15, labelpad=15)  #labels
plt.title("India's Electric Vehicle Acceleration", weight=200, family='sans-serif', size=15, pad=15)  #title
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(['Acceleration'], loc='upper right')  #legend
plt.show()
```

```python
ds2=pd.read_csv("EV_Sales.csv")
plt.figure(figsize=(20, 6))
plt.plot(ds2['YEAR'], ds2['BUS'], marker='o', linestyle='-', color='violet')
plt.xlabel('Brands', family='serif', fontsize=15, labelpad=15)
plt.ylabel('Acceleration (Sec)', family='serif', fontsize=15, labelpad=15)
plt.title("India's Electric Vehicle Acceleration", weight=200, family='sans-serif', size=15, pad=15)
plt.xticks(rotation=45)
plt.tight_layout()
plt.legend(['Acceleration'], loc='upper right')
plt.show()
```
